```python
sales_gen = df.groupby(['Age'], as_index=False)['Social Media Usage (Hours per day)'].sum().sort_values(by='Social Media Usage (Hours per day)', ascending=False)

sns.barplot(x = 'Age',y= 'Social Media Usage (Hours per day)' ,data = sales_gen)
```

```python
sales_gen = df.groupby(['Diet Quality'], as_index=False)['Mental Stress Level'].sum().sort_values(by='Mental Stress Level', ascending=False)

sns.barplot(x = 'Diet Quality',y= 'Mental Stress Level' ,data = sales_gen)
```

```python
plt.figure(figsize=(5,6))
sales_gen = df.groupby(['Medical Condition'], as_index=False)['Mental Stress Level'].sum().sort_values(by='Mental Stress Level', ascending=False)

sns.barplot(x = 'Medical Condition',y= 'Mental Stress Level' ,data = sales_gen)
plt.title('Comparison of medical condition with stressed people')
```

```python
q5 = df.groupby("Stress Coping Mechanisms").agg({"Gender":"count", "Age":"mean"}).reset_index()
```

```python
fig, ax1 = plt.subplots(figsize=(25
                                 , 5))

ax1.bar(q5["Stress Coping Mechanisms"], q5["Gender"], color='skyblue', label="Car Count")
ax1.set_ylabel("Stress Coping Mechanisms by gender", color='black')
ax1.tick_params(axis='y', labelcolor='black')

ax2 = ax1.twinx()
ax2.plot(q5["Stress Coping Mechanisms"], q5["Age"], color='red', marker='o', label="Avg Price")
ax2.set_ylabel("Stress Coping Mechanisms by age", color='red')
ax2.tick_params(axis='y', labelcolor='red')

plt.title("Stress Coping Mechanisms by Gender and Age")
plt.xlabel("Stress Coping Mechanisms")
plt.ylabel("Count")
plt.legend(loc="upper left")
plt.tight_layout()
plt.show()
```
